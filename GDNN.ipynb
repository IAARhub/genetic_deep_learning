# Genetic Deep Neural Network

Parsing data from safirortb.log

```python
import numpy as np
import os
import math

path = os.getcwd()
path 
gdnn = np.loadtxt(path + "/results/GDNN_one_layer.csv", delimiter=",")
dnn = np.loadtxt(path + "/results/DNN_one_layer.csv", delimiter=",")

dnn[:,1]
```

```python
from matplotlib.pyplot import figure
import matplotlib.pyplot as plt
%matplotlib inline

figure(num=None, figsize=(8, 4), dpi=80, facecolor='w', edgecolor='k')

plt.plot(gdnn[:,0],gdnn[:,1], label="GDNN")
plt.plot(dnn[:,0],dnn[:,1], label="DNN")

plt.title('GDNN vs DNN loss through generations')
plt.xlabel('epochs (600 epochs is a new generation for GDNN)')
plt.ylabel('loss');
plt.legend()
plt.show()
```
